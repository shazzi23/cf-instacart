## 3.1 If-Statements with User-Defined Functions

```python
# Define function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the user_defined variable to the dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

## 3.3 If-Statements with For-Loops

```python
# Identify busiest day of the week for orders based on counts
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create loop based on busiest day of the week for orders.
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Add a new column to dataframe for the busiest day of the week and use result list for values
ords_prods_merge['busiest_day'] = result
```

## 4.2 Create column based on busiest time period.

```python
# Check column names to see which column to use for my 'busiest_time_period' column.
list(ords_prods_merge.columns.values)
```

```python
# Identify busiest time period for orders based on counts
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

__Evaluation:__ The above will be split into three even groups according to the below"
- "Most orders" includes the following hours: [10, 11, 14, 15, 13, 12, 16, 9]
- "Fewest orders" includes the following hours: [3, 4, 2, 5, 1, 0, 6, 23] 
- "Average orders" includes the following hours: everything else

```python
# Create loop based on busiest period of day for orders.
result_3 = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [3, 4, 2, 5, 1, 0, 6, 23] :
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
# Add a new column to dataframe for the busiest time periods and use result_2 list for values
ords_prods_merge['busiest_period_of_day'] = result_3
```

```python
# View frequency of 'busiest_period_of_day'
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
